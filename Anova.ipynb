```python
import pandas
import matplotlib.pyplot as plt
import scipy

#a hypoothesis is a claim on a given dataset
#we use statistical methods to prove the claim-true/false

df = pandas.read_csv("school.csv")
print(df.isnull().sum())

#sprint is dependent - numeric
#smoking is independent - categorical
#H0: µ1= µ2= µ3= ... = µk("all k population means are equal")
#Ho: grp1..grp2..grpk(all means are equal)
#H1: grp1...grp2..grpk(atleast one mean is different from others)

medianSprint = df['Sprint'].median()
df['Sprint'].fillna(medianSprint, inplace=True)

df['Smoking'].fillna(3, inplace = True)
df['Smoking'].replace({0: 'Non', 1:'Past', 2:'Current', 3: 'Unknown'})

#we use statsmodel not scipy
from statsmodels.formula.api import ols #ordinary least squares

results = ols('Sprint ~ C(Smoking)', data=df).fit()
print(results.summary())
print('==================================')

import reserchpy
print(reserchpy.summary_count(df['Sprint'].groupby(df['Smoking'])))

# we reject the null hypothesis and take alternative hypothesis
#we conclude atleast one is different
#paired ttest -2 numeric varibales e.g math and eng
#Anova- 1 numeric (i.e numeric) with respect to categorical i.e smoking

#chi square test for independence- require 2 categorical variables ie gender and smoking
#h0, gender is not associated with smoking
#h1 gender is asociated with smoking

#df['Gender'].replace({0: 'Male', 1:'Female'})

#alpha = 0.85
#print(results)
#if 0.3597 < alpha:
#    print('Reject Null, Take alternative; Gender is associated wwith smoking')
#else:
#    print('Accept Null, Gender is not associated with smoking')
```
