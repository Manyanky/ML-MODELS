```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn

from sklearn import model_selection
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score, mean_squared_error


# Loading the Train and Test Sets
Training_data = pd.read_csv('Train_maskedv2.csv')
Testing_data = pd.read_csv('Test_maskedv2.csv')
full_data = [Training_data, Testing_data]

Training_data.head(2)
```

```python
#specifying our target variable

target = ['target_pct_vunerable']
```

```python
#extracting columns we want to use into a new dataframe ,features

features = Training_data[['total_households', 'total_individuals','dw_00', 'dw_01', 'dw_02', 'dw_03','dw_04', 'dw_05', 'dw_06', 'dw_07', 'dw_08', 'dw_09','dw_10',
       'dw_11', 'dw_12','dw_13','psa_00', 'psa_01', 'psa_02','psa_03', 'psa_04', 'stv_00', 'stv_01', 'car_00',
       'car_01','lln_00', 'lln_01', 'lan_00', 'lan_01','lan_02', 'lan_03', 'lan_04','lan_05','lan_06','lan_07', 'lan_08',
       'lan_09', 'lan_10','lan_11', 'lan_12', 'lan_13', 'lan_14','pg_00', 'pg_01', 'pg_02', 'pg_03', 'pg_04', 'lgt_00']]
```

```python
X = features
Y = Training_data[target]
```

```python

#Training_data.drop(['ward'],inplace=True,axis=1)
#Testing_data.drop(['ward'],inplace=True,axis=1)

#converting columns to float
#Training_data[1:19] = Training_data[1:19].astype(float)
#Testing_data[1:19] = Testing_data[1:19].astype(float)
```

```python
# Checking the missing values

missing_df = Training_data.isnull().sum(axis=0).reset_index()
missing_df.columns = ['Column Name', 'Missing Values Count']
missing_df['Filling Factor (%)']=(Training_data.shape[0]-missing_df['Missing Values Count'])/Training_data.shape[0]*100
missing_df.sort_values('Filling Factor (%)').reset_index(drop = True)
```

```python
#splitting data to X and Y
#array = Training_data.values
#X = array[:, 1:49] #upto 18
#Y = array[:, 49]#target variable to be predicted
#Y = Y.astype('float')
#print(Y)
```

```python

X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X,Y, test_size=0.30, random_state=20)
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=324)


#picking one model
model = LinearRegression()
model.fit(X_train, Y_train)  #training process

#ask the model to predict the x test while hiding the y test from the model (unseen data)
#predictions = model.predict(X_test)
predictions = model.predict(X_test)

#in regression we always check the R squared and, absolute error squared


print("R squared(%):", r2_score(Y_test, predictions))
print("mean squared error:", mean_squared_error(Y_test, predictions))
```
